## credit card fraud detection

```python
#importing dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

## data collection


```python
#loding dataset
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
#printing first 5 rows
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
#getting info about data
credit_card_data.info()
```

```python
#checking missing values
credit_card_data.isnull().sum()
```

```python
#distribution of legit transaction and fraudial transaction
credit_card_data['Class'].value_counts()
```

this dataset is highly imbalance

0->normal transaction
1->fraudial transaction

```python
#separating the data for analysis
legit = credit_card_data[credit_card_data['Class']==0]
fraud = credit_card_data[credit_card_data['Class']==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#stastical measure of data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transaction
credit_card_data.groupby('Class').mean()
```

## under sampling


built a sample dataset containing similar distribution of normal transactions and fraudtical transactions

number of fraudital transaction (492)

```python
legit_sample = legit.sample(n=492)
```

### concatenate data frames

```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

## splitting the data into feature and target

```python
X = new_dataset.drop(columns='Class',axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

## Split the data into training and testing data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

## model training

logistic regression

```python
model = LogisticRegression()
```

```python
# Check the shapes of X_train and Y_train
print("Shape of X_train:", X_train.shape)
print("Shape of Y_train:", Y_train.shape)
```

```python
#training the logistic regression model

model.fit(X_train,Y_train)
```

## model evaluation

accuracy score

```python
#accuracy the training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("accuracy on training data :",training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("accuracy score on test data :",test_data_accuracy)
```
